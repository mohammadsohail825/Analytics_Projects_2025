### Average of Tenure

**Average tenure:** The average tenure measures the average number of years an employee stays with the
company before leaving. It can provide insight into workforce stability and employee satisfaction within the
organization

```python
avg_tenure = df['YearsAtCompany'].mean()
```

```python
print(f'Average years of employee to leave the company is {avg_tenure} years')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,5))
sns.histplot(data=df, x='Age', kde=True, palette ='Set1', ax=axes[0])
axes[0].set_title('Distribution Employee by Age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Count')
sns.countplot(data=df, x='Gender',palette ='Set1', ax=axes[1], hue ='Gender')
axes[1].set_title('Distribution Employee by Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Count')
sns.countplot(data=df, x='Department', palette='Set1', ax=axes[2], hue='Department')
plt.setp(axes[2].get_xticklabels(), rotation=45)
axes[2].set_title('Distribution Employee by Department')
axes[2].set_xlabel('Department')
axes[2].set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Convert Categorical Columns to Category dtype
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    df[col] = df[col].astype('category')
```

```python
df_encoded = df.copy()
df_encoded[categorical_cols] = df_encoded[categorical_cols].apply(lambda x: x.cat.codes)
```

```python
# Make Heatmap to Showing Correlation of data
plt.figure(figsize=(14, 8))
correlation = df_encoded.corr()
sns.heatmap(correlation, annot=False, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig("eda_correlation_heatmap.png")
plt.show()
```

```python
# Make Boxplot for find Attrition acording to Age, Montly Income and total Working Years
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.boxplot(data=df, x='Attrition', y='Age', palette='Set3', hue = 'Attrition', ax=axes[0])
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', palette='Set3', hue = 'Attrition', ax=axes[1])
sns.boxplot(data=df, x='Attrition', y='TotalWorkingYears', palette='Set3', hue = 'Attrition', ax=axes[2])
axes[0].set_title('Age vs Attrition')
axes[1].set_title('Monthly Income vs Attrition')
axes[2].set_title('Total Working Years vs Attrition')
plt.tight_layout()
plt.savefig("eda_boxplots_attrition.png")
plt.show()
```

Here in Boxplot fo Attrition acording to Age, Montly Income and total Working Years. in this boxes we can see outlier in monthly income and total working years boxplot

```python
# Acording to working year make bins Entry, mid, senior and expert bise.
df['ExperienceLevel'] = pd.cut(
    df['TotalWorkingYears'],
    bins=[0, 3, 7, 15, 40],
    labels=['Entry', 'Mid', 'Senior', 'Expert']
)
```

```python
exp_counts = df['ExperienceLevel'].value_counts().sort_index()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(exp_counts, labels=exp_counts.index, autopct='%1.1f%%', startangle=140,
        colors=sns.color_palette('pastel'))
plt.title('Employee Distribution by Experience Level')
plt.axis('equal')
plt.show()
```

More than 40% of Employee are Senior and above 10% of employee entry lavel.

```python
# Acording to working year in company make bins distribution.
df["TenureGroup"] = pd.cut(
    df["YearsAtCompany"],
    bins =[0,3,7,15,40],
    labels=['0-2','4-7','8-15','15+']
)
```

#Conclusion



**​Based on the HR employee attrition analysis, several key insights emerge:​**

***Attrition Distribution:***
The dataset reveals an imbalanced distribution, with a higher number of employees staying (No) compared to those leaving (Yes).​

**Correlation Analysis:** Strong correlations exist between variables such as MonthlyIncome, TotalWorkingYears, and JobLevel, indicating that as employees gain more experience and hold higher job levels, their income increases.​

**Attrition Factors:** Employees who left the company tend to be younger, have lower income, and possess fewer total working years.​

**Experience Level Distribution:** When categorizing employees based on TotalWorkingYears into experience levels (Entry, Mid, Senior, Expert), a significant portion falls into the Entry and Mid categories, suggesting that less experienced employees are more prone to attrition.​
These findings highlight the importance of targeted retention strategies, especially for younger and less experienced employees. Addressing factors such as competitive compensation, career development opportunities, and work-life balance could be pivotal in reducing attrition rates.​
